```python
def times2(i):
    return 2 * i
df["comments2"] = df["comments"].apply(times2)
df
```

```python
from datetime import datetime
def timestamp(t):
    return str(datetime.fromtimestamp(t))
df["film_date_2"] = df["film_date"].apply(timestamp)
df
```

```python
def commentflow(i):
    if i <= 5000 and i >= 2000:
        return True
    else:
        return False
f = df["comments"].apply(commentflow)
df[f]
```

```python
def tagflow(i):
    tag_list = eval(i)
    if "children" in tag_list:
        return True
    else:
        return False
df[ df["tags"].apply(tagflow) ]
```

```python
def tagflow(i, **kwargs):
    tag_list = eval(i)
    if kwargs["target"] in tag_list:
        return True
    else:
        return False
s = input("請輸入你想要篩選的標籤:")
target_df = df[ df["tags"].apply(tagflow, target=s) ]
target_df.to_csv("filter.csv", encoding="utf-8", index=False)
target_df
```
