```python
import pandas as pd
df = pd.read_excel("example.xlsx")
df
```

```python
from rake_nltk import Rake
import nltk
import ssl
import ssl
ssl._create_default_https_context = ssl._create_unverified_context
nltk.download('stopwords')
nltk.download('punkt')
```

```python
def extract(text):
    r = Rake()
    result = r.extract_keywords_from_text(text)
    return r.get_ranked_phrases()
extract(df.iloc[0]["Title"])
```

```python
df["Title(Keyword)"] = df["Title"].apply(extract)
df["Abstract(Keyword)"] = df["Abstract"].apply(extract)
df
```

```python
df.to_csv("extract.csv", encoding="utf-8", index=False)
```
